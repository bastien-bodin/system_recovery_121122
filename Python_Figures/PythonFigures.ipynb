```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib ipympl
plt.rcParams.update({'font.size': 18})
```

```python
gamma_dot = np.arange(0.001,100,0.001)
tau_y = 10
m_pL = [0, 0.5, 1, 10]
Labels = ['Newtonian', r'$m=0.5$', r'$m=1$', r'$m=10.0$']
Linestyles = ['-', '--', '--', '--']
n_hb = 1
eta = 1
```

```python
Eta_app = []
Tau = []
for m_p in m_pL:
    eta_app = eta*gamma_dot**(n_hb-1) + tau_y/gamma_dot*(1-np.exp(-m_p*gamma_dot))
    tau = eta_app*gamma_dot
    Eta_app.append(eta_app)
    Tau.append(tau)
Eta_app = np.array(Eta_app)
Tau = np.array(Tau)
TauB = tau_y + eta*gamma_dot
```

```python
plt.figure(figsize=(7,7))
lw=3
plt.title('Description of different fluid flows')
plt.ylim(0,2.3)
plt.xlim(0,10)
plt.xlabel(r'Shear rate: $\dot \gamma$ $[s^{-1}]$')
plt.ylabel(r'Shear stress: $\tau/\tau_y$')
for i in range(len(Tau)):
    plt.plot(gamma_dot, Tau[i]/(tau_y),Linestyles[i] , label=Labels[i], linewidth=lw)
plt.plot(gamma_dot, TauB/(tau_y), ':',c='black',  label='Bingham', linewidth=lw)
plt.legend()
plt.grid()
plt.savefig('ShearStress_ShearRate.pdf', transparent=True)
```

```python
plt.figure()
for i in range(len(Tau)):
    plt.plot(Tau[i], Eta_app[i], label=Labels[i])
# plt.plot(TauB/tau_y, eta*np.ones_like(TauB),',', label='Bingham')
plt.xlabel(r'$\tau$')
plt.ylabel(r'$\eta$')
plt.xscale('log')
plt.grid()
plt.legend()
```
